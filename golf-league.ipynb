```python
import plotly.express as px
import plotly.graph_objects as go

print("Setup Complete")
```

```python
# Path of the file to read 
df2 = pd.read_csv('https://raw.githubusercontent.com/slane2014/golf_league/master/golf_temp2.csv')

# Read score sets into individual tuples
scoresets = [tuple(x) for x in df2.values]

# X-axis labels
Rounds = ['Round 1','Round 2','Round 3','Round 4','Round 5','Round 6','Round 7','Round 8','Round 9','Round 10','Round 11','Round 12']

# Create figure
fig = go.Figure()

# Add the scores for each individual player
for x in range(len(scoresets)):
    fig.add_trace(go.Scatter(x=Rounds, y=scoresets[x][1:13], name=scoresets[x][0],))

# Edit the layout
fig.update_layout(title='Scores by Round', xaxis_title='Round Number', yaxis_title='Score')
fig.show()
```
